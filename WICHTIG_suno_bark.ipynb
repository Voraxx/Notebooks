## Advanced Vocal Sample Generation



```
# Prompts for pet droid vocal samples:
pet_droid_prompt_1 = """I really. Like. You."""
```



````python
# Scary woman vocal samples:

scary_woman_prompt_1 = """
     WOMAN: [laughs] I will destroy you! [shouts] I will destroy you!
"""
audio_array = generate_audio(scary_woman_prompt_1, history_prompt="v2/en_speaker_9", waveform_temp=0.5)
Audio(audio_array, rate=SAMPLE_RATE)
````

````python
# Calm woman vocal samples:

calm_woman_prompt_1 = """
    WOMAN: Serenity and calmness.
"""
audio_array = generate_audio(calm_woman_prompt_1, history_prompt="v2/en_speaker_9", waveform_temp=0.5)
Audio(audio_array, rate=SAMPLE_RATE)
````

````python
# Scared woman vocal samples:

scared_woman_prompt_1 = """
    WOMAN: Do not go in there. You will die.
"""
audio_array = generate_audio(scared_woman_prompt_1, history_prompt="v2/en_speaker_9", waveform_temp=0.5)
Audio(audio_array, rate=SAMPLE_RATE)
````

````python
# Pet droid vocal samples:
pet_droid_prompt_1 = """I really... Like... You..."""

audio_array = generate_audio(pet_droid_prompt_1, history_prompt="v2/en_speaker_5", waveform_temp=0.5)
Audio(audio_array, rate=SAMPLE_RATE, embed="true")
````
